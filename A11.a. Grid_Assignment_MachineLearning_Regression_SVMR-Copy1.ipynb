```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset=pd.read_csv("insurance_pre.csv")
# Reading the dataset .data is Assigned to class pandas(pd) dot function name read_csv("csvfile Name.csv")
```

```python
dataset
```

```python
dataset=pd.get_dummies(dataset,drop_first=True)
```

```python
dataset
```

```python
dataset.columns
```

```python
independent=dataset[['age', 'bmi', 'children', 'sex_male', 'smoker_yes']]
#input(YearsExperience) is assigned to variable independent
#closed Double bracket used for calling one column fromthe .csv file
```

```python
dependent=dataset[["charges"]]
##Output(Salary) is assigned to variable dependent
#closed Double bracket used for calling one column fromthe .csv file
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(independent,dependent,test_size=0.20,random_state=0)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVR
param_grid ={'kernel':['rbf','poly','sigmoid','linear'],'C':[10,100,1000,2000,3000],'gamma':['auto','scale']}
grid = GridSearchCV(SVR(),param_grid,refit= True,verbose=3,n_jobs=-1)
#fitting the model for grid search
grid.fit(independent,dependent)
```

```python
#print best parameter after tuning
#print(grid.best_params_)
re=grid.cv_results_
print("The R_score value for best parameter {}:".format(grid.best_params_))
```

```python
table=pd.DataFrame.from_dict(re)
```
